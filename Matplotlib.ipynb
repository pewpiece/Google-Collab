```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
url = 'https://raw.githubusercontent.com/dphi-official/Datasets/master/Standard_Metropolitan_Areas_Data-data.csv'
df = pd.read_csv(url)

# Clean column names
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')

# Calculate correlation
corr = df['hospital_beds'].corr(df['physicians'])
print(f"Correlation between hospital_beds and physicians: {corr:.2f}")

# Plotting with matplotlib only
plt.figure(figsize=(8, 6))
plt.scatter(df['hospital_beds'], df['physicians'], color='blue', alpha=0.6, label='Data points')

# Best fit line (manually using numpy)
m, b = pd.Series(df['hospital_beds']).cov(df['physicians']) / df['hospital_beds'].var(), \
       df['physicians'].mean() - (df['hospital_beds'].mean() * pd.Series(df['hospital_beds']).cov(df['physicians']) / df['hospital_beds'].var())
plt.plot(df['hospital_beds'], m * df['hospital_beds'] + b, color='red', label='Best fit line')

plt.title(f'Hospital Beds vs Physicians\nCorrelation: {corr:.2f}')
plt.xlabel('Hospital Beds')
plt.ylabel('Physicians')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

#  Lesson 1: Introduction to Matplotlib & Dataset Overview

#  Lesson 2: Basic Plotting with Matplotlib
2.1. Histogram of Math Scores
Histograms are great for understanding the distribution of a single variable. Let's plot the distribution of math scores:

```python
plt.hist(df['math score'], bins=10, color='skyblue', edgecolor='black')
plt.title('Distribution of Math Scores')
plt.xlabel('Math Score')
plt.ylabel('Number of Students')
plt.show()
```

2.2. Bar Plot of Average Scores by Gender
Bar plots can compare averages across categories. Let's compare average scores between genders:

```python
avg_scores = df.groupby('gender')[['math score', 'reading score', 'writing score']].mean()
avg_scores.plot(kind='bar')
plt.title('Average Scores by Gender')
plt.ylabel('Average Score')
plt.xticks(rotation=0)
plt.show()
```

## Lesson 3: Advanced Visualizations

 Box plots show the distribution and identify outliers. Let's see how test preparation affects scores:

Scatter plots reveal relationships between two variables. Let's examine reading and writing scores:

```python
sns.scatterplot(x='reading score', y='writing score', hue='gender', data=df)
plt.title('Reading vs. Writing Scores')
plt.show()
```

Matplotlib allows extensive customization. Let's enhance our previous scatter plot:

## Lesson 7: Time Series and Line Plots
Although our dataset isn't time-based, if we had time-series data, we could plot trends over time:

To work with time series or line plots, you'd need an actual date or time-related column in the dataset. Since this dataset doesn't include one, we can simulate a timeline if needed  for example, by assigning fake dates assuming the rows are in chronological order.

 Example: Simulate a date range

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
url = 'https://raw.githubusercontent.com/dphi-official/Datasets/master/exam_scores.csv'
df = pd.read_csv(url)

# Simulate dates assuming each row is a daily record
df['date'] = pd.date_range(start='2024-01-01', periods=len(df), freq='D')

# Plot math score over time
df.set_index('date')['math score'].plot(figsize=(10, 5), color='orange')
plt.title('Simulated Math Scores Over Time')
plt.xlabel('Date')
plt.ylabel('Math Score')
plt.grid(True)
plt.show()
```

## Lesson 8: Combining Multiple Plots
We can display multiple plots side by side for comparison:

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
sns.histplot(df['reading score'], bins=10, ax=axes[0], color='green')
axes[0].set_title('Reading Score Distribution')
sns.histplot(df['writing score'], bins=10, ax=axes[1], color='purple')
axes[1].set_title('Writing Score Distribution')
plt.tight_layout()
plt.show()
```

```python
# Compute correlation matrix
corr = df[['math score', 'reading score', 'writing score']].corr()

# Plot
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Between Exam Scores')
plt.show()
```
